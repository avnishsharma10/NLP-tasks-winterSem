```python
import nltk

texts="""Christopher Hemsworth[1] (born 11 August 1983)[2] is an Australian actor. He rose to prominence playing Kim Hyde in the Australian TV series Home and Away (2004–07) before beginning a film career in Hollywood by taking on parts in the science fiction film Star Trek (2009) and the thriller A Perfect Getaway (2009)."""
```

```python
text
```

```python
from nltk.tokenize import word_tokenize

words=word_tokenize(text)
words
```

```python
from nltk.tokenize import sent_tokenize

sents=sent_tokenize(text)
sents
```

```python
tagged_words=nltk.pos_tag(words)
```

```python
tagged_words
```

```python
for text in texts:
    sentences=nltk.sent_tokenize(text)
    for sentence in sentences:
        words=nltk.word_tokenize(sentence)
        tagged_words=nltk.pos_tag(words)
        print(tagged_words)
```
